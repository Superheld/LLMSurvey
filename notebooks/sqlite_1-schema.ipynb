# Datenbanksetup SQLite

```python
import sqlite3

# Connect to DB
connection = sqlite3.connect('../data/survey.db')
cursor = connection.cursor()
```

```python
# Define schema

cursor.execute('''
    CREATE TABLE IF NOT EXISTS questions (
               q_id INTEGER PRIMARY KEY AUTOINCREMENT,
               q_label TEXT NOT NULL,
               q_text TEXT NOT NULL,
               q_block TEXT NOT NULL
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS strategies (
               s_id INTEGER PRIMARY KEY AUTOINCREMENT,
               s_name TEXT NOT NULL,
               s_system_prompt TEXT NOT NULL,
               s_conversation_prompt TEXT NOT NULL
    )
''')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS responses (
               r_id INTEGER PRIMARY KEY AUTOINCREMENT,
               r_strat_id INTEGER NOT NULL,
               r_quest_id INTEGER NOT NULL,
               r_answer INTEGER NOT NULL,
               FOREIGN KEY (r_strat_id) REFERENCES strategies(s_id),
               FOREIGN KEY (r_quest_id) REFERENCES questions(q_id)
    )
''')
```

```python
# commit
connection.commit()
connection.close()
```
