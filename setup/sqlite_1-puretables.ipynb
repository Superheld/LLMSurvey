# Setup Tables

Hier werden nur noch jene Tabellen angelegt, deren Inhalt noch nicht bekannt ist, praktisch keine Stammdaten sind. Tabellen mit Inhalt, wie die Fragen oder den Models, werden im entsprechenden Nodebook erstellt.

```python
import sqlite3

# Connect to DB
connection = sqlite3.connect('../data/survey.db')
cursor = connection.cursor()
```

```python
# Erstelle run-table
cursor.execute('DROP TABLE IF EXISTS runs')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS runs (
               id INTEGER PRIMARY KEY AUTOINCREMENT,
               model_id INTEGER NOT NULL,
               strategy_id INTEGER NOT NULL,
               prompt_tokens INTEGER NOT NULL,
               completion_tokens INTEGER NOT NULL,
               duration_time REAL,
               timestamp TEXT,
               FOREIGN KEY (model_id) REFERENCES models(id),
               FOREIGN KEY (strategy_id) REFERENCES strategies(id)
    )
''')
```

```python
# Create responses table
cursor.execute('DROP TABLE IF EXISTS responses')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS responses (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        run_id INTEGER NOT NULL,
        question_id INTEGER NOT NULL,
        answer INTEGER NOT NULL,
        FOREIGN KEY (run_id) REFERENCES runs(id)
    )
''')
```

```python
# commit
connection.commit()
connection.close()
```
