# Strategies Setup

6 Strategien = 3 Modi × 2 Framings

**Modi:**
- oneshot: Alle 29 Fragen auf einmal
- conversation: Frage für Frage MIT Message-History
- questionbyquestion: Frage für Frage OHNE History

**Framings:**
- none: Minimal, nur JSON-Format
- test: Test-Framing ("Du nimmst an einem Test teil")

**Prompts:**
- System Prompts: `data/prompts/system_{framing}.md`
- Conversation Templates: `data/prompts/{modus}.md`

```python
import sqlite3

# Connect to DB
connection = sqlite3.connect('../data/survey.db')
cursor = connection.cursor()

strategies = []
```

```python
# One-Shot Strategies

strategies.append({
    'name': 'oneshot_none',
    'system_prompt': 'data/prompts/system_none.md',
    'conversation_prompt': 'data/prompts/oneshot.md'
})

strategies.append({
    'name': 'oneshot_test',
    'system_prompt': 'data/prompts/system_test.md',
    'conversation_prompt': 'data/prompts/oneshot.md'
})
```

```python
# Conversation Strategies (with message history)

strategies.append({
    'name': 'conversation_none',
    'system_prompt': 'data/prompts/system_none.md',
    'conversation_prompt': 'data/prompts/conversation.md'
})

strategies.append({
    'name': 'conversation_test',
    'system_prompt': 'data/prompts/system_test.md',
    'conversation_prompt': 'data/prompts/conversation.md'
})
```

```python
# Question-by-Question Strategies (isolated, no history)

strategies.append({
    'name': 'questionbyquestion_none',
    'system_prompt': 'data/prompts/system_none.md',
    'conversation_prompt': 'data/prompts/questionbyquestion.md'
})

strategies.append({
    'name': 'questionbyquestion_test',
    'system_prompt': 'data/prompts/system_test.md',
    'conversation_prompt': 'data/prompts/questionbyquestion.md'
})
```

```python
# Prepare insert data
insert = []

for strategy in strategies:
    insert.append((
        strategy['name'],
        strategy['system_prompt'],
        strategy['conversation_prompt']
    ))

# Check count
print(f"Inserting {len(insert)} strategies")
for s in strategies:
    print(f"  - {s['name']}")
```

```python
cursor.executemany(
    "INSERT INTO strategies (name, system_prompt, conversation_prompt) VALUES (?,?,?)",
    insert
)
```

```python
connection.commit()
connection.close()
```
