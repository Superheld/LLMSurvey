```python
import sqlite3

# Connect to DB
connection = sqlite3.connect('../data/survey.db')
cursor = connection.cursor()

models = []
```

```python
# Create models table
cursor.execute('DROP TABLE IF EXISTS models')

cursor.execute('''
    CREATE TABLE IF NOT EXISTS models (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        model_id TEXT NOT NULL UNIQUE,
        provider TEXT,
        name TEXT
    )
''')
```

```python
# Mistral

models.append({
    'name': 'Mistral Small',
    'provider': 'Mistral',
    'model_id': 'mistral/mistral-small-2506'
})

models.append({
    'name': 'Mistral Large',
    'provider': 'Mistral',
    'model_id': 'mistral/mistral-large-2512'
})
```

```python
# Open AI

models.append({
    'name': 'GPT-5.2 Nano',
    'provider': 'Open AI',
    'model_id': 'openai/gpt-5-nano-2025-08-07'
})

models.append({
    'name': 'GPT-5.2',
    'provider': 'Open AI',
    'model_id': 'openai/gpt-5.2-2025-12-11'
})
```

```python
# Anthropic

models.append({
    'name': 'Claude Haikaku',
    'provider': 'Anthropic',
    'model_id': 'anthropic/claude-haiku-4-5-20251001'
})

models.append({
    'name': 'Claude Opus',
    'provider': 'Anthropic',
    'model_id': 'anthropic/claude-opus-4-5-20251101'
})
```

```python
# Google

models.append({
    'name': 'Gemini Flash',
    'provider': 'Google',
    'model_id': 'gemini/gemini-3-flash-preview'
})

models.append({
    'name': 'Gemini Pro',
    'provider': 'Google',
    'model_id': 'gemini/gemini-3-pro-preview'
})
```

```python
# Deepseek

models.append({
    'name': 'Deepseek Chat',
    'provider': 'Deepseek',
    'model_id': 'deepseek/deepseek-chat'
})

models.append({
    'name': 'Deepseek Reasoner',
    'provider': 'Deepseek',
    'model_id': 'deepseek/deepseek-reasoner'
})
```

```python
# Prepare insert data

insert = []

for model in models:
    insert.append((
        model['name'],
        model['provider'],
        model['model_id']
    ))
```

```python
cursor.executemany(
    "INSERT INTO models (name, provider, model_id) VALUES (?,?,?)",
    insert
)
```

```python
connection.commit()
connection.close()
```
